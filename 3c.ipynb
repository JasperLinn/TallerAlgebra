```python

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Vectores a graficar
vector1 = np.array([2, 4, 12])
vector2 = np.array([14, 28, 96])

# Configurar el gráfico 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Graficar los vectores
ax.quiver(0, 0, 0, vector2[0], vector2[1], vector2[2], color='c', label='Vector 2')
ax.quiver(0, 0, 0, vector1[0], vector1[1], vector1[2], color='y', label='Vector 1')

# Etiquetar los ejes
ax.set_zlabel('Eje z')
ax.set_ylabel('Eje Y')
ax.set_xlabel('Eje x')

# Configurar los límites de los ejes
ax.set_xlim([0, np.max([vector1[0], vector2[0]])])
ax.set_ylim([0, np.max([vector1[1], vector2[1]])])
ax.set_zlim([0, np.max([vector1[2], vector2[2]])])

# Mostrar leyenda y gráfico
ax.legend()
plt.show()
```
